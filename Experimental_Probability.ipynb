# Law of Large Numbers

Simply states that if a probability experiment is conducted enough times the experimental result will match the expected value from the theoritical calculations. 

Below are some simple experiments verifying this result...

```python
# Experimental Results for Coin Flip
import matplotlib.pyplot as plt
import numpy as np
numOfTrials = 1000000
results = [None] * numOfTrials

for i in range(numOfTrials):
  trialResult = np.random.randint(0,2)
  results[i] = trialResult

probOfHeads = results.count(1)/numOfTrials
probOfTails = results.count(0)/numOfTrials

plt.bar(['Heads', 'Tails'], [probOfHeads, probOfTails])
```

```python
# Experimental Result for Dice Rolls
import matplotlib.pyplot as plt
import numpy as np
numOfTrials = 1000000

results = [None] * numOfTrials

for i in range(numOfTrials):
  trialResult = np.random.randint(1,7)
  results[i] = trialResult


probOfOne = results.count(1)/numOfTrials
probOfTwo = results.count(2)/numOfTrials
probOfThree = results.count(3)/numOfTrials
probOfFour = results.count(4)/numOfTrials
probOfFive = results.count(5)/numOfTrials
probOfSix = results.count(6)/numOfTrials

plt.bar([1, 2, 3, 4, 5 , 6], [probOfOne, probOfTwo, probOfThree, probOfFour, probOfFive, probOfSix])

```

```python
# Experimental Result for Sum of Dice Rolls
import matplotlib.pyplot as plt
import numpy as np
numOfTrials = 1000000

results = [None] * numOfTrials

for i in range(numOfTrials):
  trialResultofDice1 = np.random.randint(1,7)
  trialResultofDice2 = np.random.randint(1,7)
  results[i] = trialResultofDice1 + trialResultofDice2

probOfTwo = results.count(2)/numOfTrials
probOfThree = results.count(3)/numOfTrials
probOfFour = results.count(4)/numOfTrials
probOfFive = results.count(5)/numOfTrials
probOfSix = results.count(6)/numOfTrials
probOfSeven = results.count(7)/numOfTrials
probOfEight = results.count(8)/numOfTrials
probOfNine = results.count(9)/numOfTrials
probOfTen = results.count(10)/numOfTrials
probOfEleven = results.count(11)/numOfTrials
probOfTwelve = results.count(12)/numOfTrials

plt.bar(range(2,13), [probOfTwo, probOfThree, probOfFour, probOfFive, probOfSix, probOfSeven, probOfEight, probOfNine, probOfTen, probOfEleven, probOfTwelve])
```

```python
# Experimental Results of Probaility of pulling a Red or Blue marble from a box consisting of 100 Blue and 50 Red marbles
import matplotlib.pyplot as plt
import numpy as np
numOfTrials = 1000000

results = [None] * numOfTrials
for i in range(numOfTrials):
  trialResult = np.random.randint(1,151)
  if trialResult <= 50:
    results[i] = 'r'

  else:
    results[i] = 'b'

probOfRed = results.count('r')/numOfTrials
probOfBlue = results.count('b')/numOfTrials


plt.bar(['Blue', 'Red'], [probOfBlue, probOfRed])
print("The probability of pulling a Blue Marble is: " + str(probOfBlue))
print("The probability of pulling a Red Marble is: " + str(probOfRed))
```

```python
# Experimental results for the 'Birthday Problem'
%%time
import matplotlib.pyplot as plt
import numpy as np


numOfTrials = 10000

def genBirthday(numOfBirthdays):
  birthdays = []
  for birthday in range(numOfBirthdays):
    birthdays.append(np.random.randint(1,366))

  return birthdays

def isDuplicate(birthdays):
  if len(birthdays) == len(set(birthdays)):
    return False
  else:
    return True


results = []

for numOfPeople in range(100):
  trialResult = []
  for i in range(numOfTrials):
    birthdays = genBirthday(numOfPeople)
    birthdays.sort()
    
    if isDuplicate(birthdays) == True:
      trialResult.append(1)
    else:
      trialResult.append(0)
  
  results.append(trialResult.count(1)/numOfTrials)


plt.xlabel('# of People')
plt.ylabel('Probability')
plt.title("Probability of the same birthday")
ax = plt.gca()
ax.plot(np.arange(1, 101), results)
ax.plot(23, results[22], 'b', marker="o",markeredgecolor="red", markerfacecolor="green")
plt.show()
```

```python
# Experimental results for approximating the value of Pi via darts

import matplotlib.pyplot as plt
import numpy as np


numOfTrials = 1000
min_X = 0
max_X= 1
min_Y = 0
max_Y = 1

plt.title("Random Darts to Pi")
ax = plt.gca()

x = np.arange(0 ,1, 0.005)
y= np.sqrt(1 - x**2)


ax.plot(x,y)


results = [None] * numOfTrials

for i in range(numOfTrials):
  dart_X = (max_X - min_X) * np.random.random_sample() + min_X
  dart_Y = (max_Y - min_Y) * np.random.random_sample() + min_Y

  if dart_X**2 + dart_Y**2 <= 1:
    ax.plot(dart_X, dart_Y, 'g', marker="x")
    results[i] = 1

  else:
    ax.plot(dart_X, dart_Y, 'r', marker="x")
    results[i] = 0
print(results)
pi = 4 * results.count(1)/numOfTrials
plt.show()
print(pi)
```
